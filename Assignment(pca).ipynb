```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('wine.csv')
data.head()
```

```python
data2=data.iloc[:,1:]
data2
```

```python
  data2.shape
```

```python
data2.info()
```

```python
data_ary = data2.values
data_ary
```

```python
# normalization of data
n_data= scale(data_ary)
n_data
```

```python
#performing pca
pca= PCA(n_components=13)

pca_values= pca.fit_transform(n_data)
pca_values
```

```python
# loadings or weights
pca.components_
```

```python
# the amount of variance in pca components
var=pca.explained_variance_ratio_
var
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained
plt.plot(var1,color='blue')
```

```python
# first three pca components has more coverage
#as per the qst data set
final_df=pd.concat([data['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# visualization pca
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
# 1) hierarchial clustering
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
# we already normalized our data
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(n_data,'complete'))
```

```python
# creating cluster
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(n_data),columns=['clusters'])
y
```

```python
data3=data.copy()
data3['H_clusters']=y
data3
```

```python
# k means clustering
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# data alres=ady normalized
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 6):
    clf = KMeans(n_clusters=i)
    clf.fit(n_data)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1,6), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
# cluster alg with k=3
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(n_data)
```
